```python
#importing packages

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#loding an image
image = cv2.imread("/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/openCVLogo.png")
```

```python
plt.imshow(image)
```

```python
#it is displaying in BGR we have to convert it into RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#we have to define affine transform matrix
#translating an image in numpy martix is given by
#       [1, 0, shiftX]
#       [0, 1, shiftY]
# negative values of X will shift the image to the left
# negative values of Y will shift the image to the up
# positive values of X will shift the image to the right
# negative values of Y will shift the image to the down

#shift 25 pixels to the right and 50 pixels down
M = np.float32([[1,0, 25],[0,1,50]])

#apply cv2.warpAffine function to shift
shifted = cv2.warpAffine(image, M, (image.shape[1],image.shape[0]))
```

```python
plt.imshow(shifted)
```

```python
M = np.float32([[1,0,-25],[0,1,-50]])
shifted = cv2.warpAffine(image, M, (image.shape[1],image.shape[0]))
```

```python
plt.imshow(shifted)
```
