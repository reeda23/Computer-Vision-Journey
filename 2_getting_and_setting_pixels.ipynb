```python
#imprting packages to be used
import matplotlib.pyplot as plt
import cv2
```

```python
#loading image 

image = cv2.imread("/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/persianCat.jpg")
```

```python
(h,w) = image.shape[:2]
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title("Persian Cat")
plt.imshow(image)
```

```python
# Consider images as numpy arrays
# Origin of the image is at (0,0) which is the top left most corner

(b,g,r) = image[0,0]
print("Pixel at (0, 0) - Red {}, Green {}, Blue {}".format(r,g,b))
```

```python
#accessing the pixels located at x = 50 and y = 20 
(b,g,r) = image[20,50]  # we write this as 20 first and 50 latter because y are the rows and x are the cololumns 

print("Pixel at (50, 20) - Red {}, Green {}, Blue {}".format(r,g,b))
```

```python
image[20,50] = (0,0,255)
(b, g , r) = image[20,50]
```

```python
plt.title("Cat")
plt.imshow(image)
```

```python
print("Pixel at (50, 20) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
```

```python
#computing the center of the image 

(cX, cY) = (w//2, h//2)
```

```python
#extracting top left corner of the image

tl = image[0:cY, 0:cX]
plt.imshow(tl)
plt.title("Top left corner")
```

```python
tr = image[0:cY, cX:w]
plt.imshow(tr)
plt.title("Top right corner")
```

```python
bl = image[cY:h, 0:cX]
plt.imshow(bl)
plt.title("Bottom left corner")
```

```python
br = image[cY:h, cX:w]
plt.imshow(br)
plt.title("Bottom right corner")
```
