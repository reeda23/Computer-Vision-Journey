```python
#importing libraries 

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load an image from the disk
image = cv2.imread('/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/persianCat.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title('Original Image')
```

```python
#reize the image to 150px wide
#for this we have to calculate the aspect ratio to prevent the image to be distorted
#for this calculate the ratio of *new* width to the *old* width

r = 150 / image.shape[1]
dim = (150, int(image.shape[0] *r))

#performing the actual resizing
resized_image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_image)
plt.title('image')
```

```python
#resize the image to 50px height

r = 50 / image.shape[0]
dim = (int(image.shape[1]), 50)

resized_image_h = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_image_h)
plt.title("Resized image with respect to height")
```

```python
def resize(image, width=None, height=None, inter = cv2.INTER_AREA):

    #get the image dimenstion
    dim = None
    (h,w) = image.shape[:2]

    #if the width and height both are None then return original image
    if width is None and height is None:
        return image

    #if the width is None

    if width is None:
        #calculate the ratio of new height and old and get the dimensions

        r = height/ float(h)
        dim = (int(w*r), height)

    #otherwise height is None
    else:
        r = width/ float(w)
        dim = (width, int(h*r))

    resized = cv2.resize(image, dim, interpolation = inter)

    #return the resized image
    return resized
```

```python
scaling_image = resize(image, width=10)
plt.imshow(scaling_image)
```

```python
scaling_image = resize(image, width = 100, height=100)
plt.imshow(scaling_image)
```

```python
#trying all the interpolations in cv2

methods = [
           ("cv2.INTER_NEAREST", cv2.INTER_NEAREST),
           ("cv2.INTER_LINEAR", cv2.INTER_LINEAR),
           ("cv2.INTER_AREA",cv2.INTER_AREA),
           ("cv2.INTEE_CUBIC",cv2.INTER_CUBIC),
           ("cv2.INTER_LANCZOS4",cv2.INTER_LANCZOS4)
]


for (name, method) in methods:
    # increase the size of the image by 3x using the current
    # interpolation method
    print("[INFO] {}".format(name))
    scaled_image = resize(image, width=image.shape[1] * 3, height= 50, inter=method)
    plt.figure() #write this otherwise plt.imshow() will not display all the images in a loop
    plt.imshow(scaled_image)
    plt.title(name)
```
