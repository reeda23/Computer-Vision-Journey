```python
#import packages
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#for drawing a rectangle

canvas_r = np.zeros((300,300), dtype="uint8")
canvas_r = cv2.cvtColor(canvas_r, cv2.COLOR_BGR2RGB)
rectangle = cv2.rectangle(canvas_r, (25,25),(275,275),(255,255,255),-1)
plt.imshow(rectangle)
plt.title("Rectangle")
```

```python
#canvas for circle

canvas_c = np.zeros((300,300), dtype="uint8")
canvas_c = cv2.cvtColor(canvas_c, cv2.COLOR_BGR2RGB)
circle = cv2.circle(canvas_c, (150,150), 150, (255,255,255),-1)
plt.imshow(circle)
plt.title("Circle")
```

```python
#bitwise AND
bitwiseAnd = cv2.bitwise_and(rectangle, circle)
bitwiseAnd = cv2.cvtColor(bitwiseAnd, cv2.COLOR_BGR2RGB)
plt.imshow(bitwiseAnd)
plt.title("BITWISE AND")
```

```python
#bitwise OR
bitwiseOr = cv2.bitwise_or(rectangle, circle)
bitwiseOr = cv2.cvtColor(bitwiseOr, cv2.COLOR_BGR2RGB)
plt.imshow(bitwiseOr)
plt.title("BITWISE OR")
```

```python
#bitwise XOR
bitwiseXOR = cv2.bitwise_xor(rectangle, circle)
bitwiseXOR = cv2.cvtColor(bitwiseXOR, cv2.COLOR_BGR2RGB)
plt.imshow(bitwiseXOR)
plt.title("BITWISE XOR")
```

```python
#bitwise Not
bitwiseNot = cv2.bitwise_not(rectangle)
plt.imshow(bitwiseNot)
plt.title("BITWISE NOT")
```
