```python
#Import packages
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#read image and convert it into RGB
image = cv2.imread('/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/adrian.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Image")
```

```python
#mask has same size as input image but it has only two values 0 and 255 (0 as background and 255 as foreground)
mask = np.zeros(image.shape[:2], dtype = "uint8")
r_mask = cv2.rectangle(mask, (0,90),(290,450),(255,255,255), -1)
r_mask = cv2.cvtColor(r_mask, cv2.COLOR_BGR2RGB)
plt.imshow(r_mask)
```

```python
#apply mask on the original image 
masked = cv2.bitwise_and(image, r_mask)
plt.imshow(masked)
plt.title("Mask applied to the image")
```

```python
#now apply mask to the face 

mask = np.zeros(image.shape[:2], dtype="uint8")
f_mask = cv2.circle(mask, (145,200), 100, (255,255,255),-1)
f_mask = cv2.cvtColor(f_mask, cv2.COLOR_BGR2RGB)
plt.imshow(f_mask)
plt.title("Mask for cropping out face")
```

```python
#apply mask on the original image 
masked = cv2.bitwise_and(image, f_mask)
plt.imshow(masked)
plt.title("Mask applied to the image")
```
