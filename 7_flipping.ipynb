```python
#importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the image

image = cv2.imread("/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/opencvlogo.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Orignal Image")
```

```python
#flipping image horizontally

flipped_h = cv2.flip(image, 1) #for horizontal flip we pass 1
plt.imshow(flipped_h)
plt.title("Flipped Image Horizontally")
```

```python
#flipping image vertically

flipped_v = cv2.flip(image, 0) #for vertical flip we pass 0
plt.imshow(flipped_v)
plt.title("Flipped Image Vertically")
```

```python
#flipping image along both axis

flipped = cv2.flip(image, -1) #for flipping along both axis we pass -1
plt.imshow(flipped)
plt.title("Flipped Image along both axis")
```
