```python
#importing packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load image 
image = cv2.imread('/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/adrian.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#display image
plt.imshow(image)
plt.title("Original Image")
```

```python
#cropping an image with openCV

#syntax cropped_image = image[startY:endY , startX:endX] | Y mean rows and X mean columns

face = image[91:250, 91:225]
plt.imshow(face)
plt.title("Face")
```

```python
body = image[250:, 0:290]
plt.imshow(body)
plt.title("Body")
```

```python
#cropping background people
bgpeople = image[181:220, 18:79]
plt.imshow(bgpeople)
plt.title("Background People")
```
