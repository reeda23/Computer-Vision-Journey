```python
#importing packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#reading an image and converting it into RGB
image = cv2.imread('/content/drive/MyDrive/OpenCV101 - Basics of OpenCV/grand_canyon.png')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#displaying original image
plt.imshow(image)
plt.title("Original Image")
```

Addition/Subtraction with OpenCV



```python
#images are stored as numpy arrays as unsigned 8 bit integers (uint8)
#when we add/subtr in openCV values will be clipped between this range i.e [0-255]

added = cv2.add(np.uint8([200]), np.uint8([100]))
subtracted = cv2.subtract(np.uint8([50]), np.uint8([100]))
print("200 + 100 = 300 but value will be clipped within 0-255")
print("Added result = {}".format(added))
print("\n50 - 100 = -50 but value will be clipped within 0-255")
print("Subtracted result = {}".format(subtracted))
```

Addition/Subtraction with Numpy

```python
#when we add/subtr in Numpy values will be wrapped between this range i.e [0-255] instead of clipped

added = np.uint8([200]) + np.uint8([100])
subtracted = np.uint8([50]) - np.uint8([100])

# will be wrapped(using modulus additon) within 0-255 like 200+100 = 300 then 300%256 = 44
print("200 + 100 = 300 but value will be wrapped within 0-255")
print("Added result = {}".format(added))

# will be wrapped(using modulus additon) within 0-255 like 50-100 = -50 then -50%256 = 206
print("\n50 - 100 = -50 but value will be clipped within 0-255")
print("Subtracted result = {}".format(subtracted))
```

```python
#increasing the intenstity of pixels in image by 100
#1- create a matrix of same dimension as our input image
#2- fill this matrix by ones
#3- multiply it by 100
#4- add it in the original image

M = np.ones(image.shape, dtype= "uint8") * 100
added = cv2.add(image, M)
plt.imshow(added)
plt.title("Lighter")
```

```python
#like addition we can subtract 50 pixels from original image and make it darker

M = np.ones(image.shape, dtype= "uint8") * 50
subtracted = cv2.subtract(image, M)
plt.imshow(subtracted)
plt.title("Darker")
```
